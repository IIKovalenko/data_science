```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:2f}'.format
```

```python
export_2015 = pd.read_excel('Russia-2015.xlsx', sheet_name='Partner')
export_2016 = pd.read_excel('Russia-2016.xlsx', sheet_name=1)
```

```python
export = pd.concat([export_2015, export_2016])
```

```python
export.sample(10)
```

```python
export.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values('Export (US$ Thousand)', ascending=False)
```

```python
exclude = ['Europe & Central Asia', ' World']
```

```python
export_filtered = export[export['Partner Name'] != 'Europe & Central Asia']
```

```python
export_filtered.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values('Export (US$ Thousand)', ascending=False)
```

```python
exclude = ['Europe & Central Asia', ' World', 'East Asia & Pacific', 'United States Minor Outlying I']
export_filtered = export[-export['Partner Name'].isin(exclude)]
```

```python
export_pivot = export_filtered.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values(['Year', 'Export (US$ Thousand)'], ascending=[True, False])
```

```python
export_pivot[export_pivot.index.get_level_values('Year')==2015]
```

```python
export_pivot[export_pivot.index.get_level_values('Year')==2016]
```
